```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import altair as alt

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df = pd.read_csv('SBAnational.csv')
df.head(1)
```

```python
# visualizations dataframe
df['DisbursementGross'] = df['DisbursementGross'].replace('[\$,]', '', regex=True).astype(float)
df['GrAppv'] = df['GrAppv'].replace('[\$,]', '', regex=True).astype(float)
df['SBA_Appv'] = df['SBA_Appv'].replace('[\$,]', '', regex=True).astype(float)
# clean 'ApprovalFY' column by removing non-numeric characters and converting to int
df['SBA_Appv'] = df['ApprovalFY'].astype(str).str.extract('(\d+)').astype(int)
df.head(10)
```

```python
import geopandas
state = geopandas.read_file('ne_110m_admin_1_states_provinces')

# manipulate string data to match the loan dataset, so instead of US-MN, it will be MN, for Minnesota etc
state['iso_3166_2'] = state['iso_3166_2'].str.replace('US-', '')

state.head(1)
```

```python
merged_init = pd.merge(df, state, left_on='State', right_on='iso_3166_2')

# selecting all columns up to and including 'region'
region_col_index = merged_init.columns.get_loc('region') + 1  # +1 to include 'region' column itself
merged_up_to_region = merged_init.iloc[:, :region_col_index]

# selecting the geometry column
merged_geometry = merged_init[['geometry']]

# concatenating the parts to get the final DataFrame
merged_final = pd.concat([merged_up_to_region, merged_geometry], axis=1)

# dropping final rows that are not needed now that data is aggregated
merged_final = merged_final.drop(columns=['hasc_maybe', 'code_hasc', 'type_en', 'name_local', 'name_alt', 'adm0_sr', 'iso_a2', 
                                          'wikipedia', 'diss_me', 'scalerank', 'featurecla', 'note', 'code_local'])

merged_final.head(1)
```

```python
# making industry column with first two digits of NAICS
merged_industry = merged_final.copy()
merged_industry['NAICS'] = merged_industry['NAICS'].astype(str)
merged_industry['Industry'] = merged_industry['NAICS'].str[:2]


merged_industry.head(1)
```

```python
# moving column to next to NAICS
columns = list(merged_industry.columns)

columns.remove('Industry')

target_index = columns.index('NAICS') + 1  
columns.insert(target_index, 'Industry')

merged_industry = merged_industry[columns]

merged_industry.head(1)
```

```python
# map NAISC to industry

industry_map = {
  '11': 'Ag/For/Fish/Hunt',
  '21': 'Min/Quar/Oil_Gas_ext',
  '22': 'Utilities',
  '23': 'Construction',
  '31': 'Manufacturing',
  '32': 'Manufacturing',
  '33': 'Manufacturing',
  '42': 'Wholesale_trade',
  '44': 'Retail_trade',
  '45': 'Retail_trade',
  '48': 'Trans/Ware',
  '49': 'Trans/Ware',
  '51': 'Information',
  '52': 'Finance/Insurance',
  '53': 'RE/Rental/Lease',
  '54': 'Prof/Science/Tech',
  '55': 'Mgmt_comp',
  '56': 'Admin_sup/Waste_Mgmt_Rem',
  '61': 'Educational',
  '62': 'Healthcare/Social_assist',
  '71': 'Arts/Entertain/Rec',
  '72': 'Accom/Food_serv',
  '81': 'Other_no_pub',
  '92': 'Public_Admin'
}

merged_industry['Industry'] = merged_industry['Industry'].map(industry_map)
```

```python
# Convert NAICS codes to numeric, setting errors='coerce' to turn invalid values into NaN
merged_industry['NAICS'] = pd.to_numeric(merged_industry['NAICS'], errors='coerce')

# Filter out entries with invalid (NaN) or 0 NAICS codes
filtered_df = merged_industry[merged_industry['NAICS'].notna() & (merged_industry['NAICS'] != 0)]

# Now, find the most common NAICS code for each state in the filtered DataFrame
most_common_naics = filtered_df.groupby('State')['NAICS'].agg(lambda x: x.value_counts().idxmax()).reset_index(name='Most_Common_NAICS')

most_common_naics['Industry'] = most_common_naics['Most_Common_NAICS'].astype(str).str[:2].map(industry_map)

most_common_naics.head()
```

```python
filtered_df['ChgOffPrinGr'].fillna(0, inplace=True)
# Ensure the relevant columns are numeric
filtered_df['GrAppv'] = pd.to_numeric(filtered_df['GrAppv'], errors='coerce')
filtered_df['DisbursementGross'] = pd.to_numeric(filtered_df['DisbursementGross'], errors='coerce')
filtered_df['ChgOffPrinGr'] = pd.to_numeric(filtered_df['ChgOffPrinGr'], errors='coerce')

# Calculate average approval, disbursement, and charge off amounts per state
avg_values = filtered_df.groupby('State').agg(
    Avg_Approval=('GrAppv', 'mean'),
    Avg_Disbursement=('DisbursementGross', 'mean'),
    Avg_ChargeOff=('ChgOffPrinGr', 'mean')
).reset_index().round(2)

# Merge the average values with the most_common_naics DataFrame
most_common_naics = pd.merge(most_common_naics, avg_values, on='State')

# Display the first few rows of the enhanced DataFrame
most_common_naics.head()
```

```python
industry_colors = {
    'Ag/For/Fish/Hunt': '#9932cc',
    'Min/Quar/Oil_Gas_ext': '#1f78b4',
    'Utilities': '#b2df8a',
    'Construction': '#33a02c',
    'Manufacturing': '#fb9a99',
    'Wholesale_trade': '#e31a1c',
    'Retail_trade': '#fdbf6f',
    'Trans/Ware': '#ff7f00',
    'Information': '#40e0d0',
    'Finance/Insurance': '#6a3d9a',
    'RE/Rental/Lease': '#ffff99',
    'Prof/Science/Tech': '#b15928',
    'Mgmt_comp': '#b5a2ca',
    'Admin_sup/Waste_Mgmt_Rem': '#9acd32',
    'Educational': '#e31a1c',
    'Healthcare/Social_assist': '#008000',
    'Arts/Entertain/Rec': '#daa520',
    'Accom/Food_serv': '#d2691e',
    'Other_no_pub': '#ffffb3',
    'Public_Admin': '#ff00ff'
}
```

```python
import folium

# Function to return the style dictionary with black borders
def get_style_function(fill_color):
    def style_function(feature):
        return {
            'fillColor': fill_color,
            'color': '#000000',  # Set border color to black
            'fillOpacity': 0.5,
            'weight': 1,  # Border line weight
        }
    return style_function

merged_gdf = state.merge(most_common_naics, left_on='iso_3166_2', right_on='State')

# create the Folium map
m = folium.Map(location=[45, -95], zoom_start=4)

industry_groups = {}
displayed_industries = set()

for _, row in merged_gdf.iterrows():
    state_geo = row['geometry']
    state_name = row['name']
    industry = row['Industry']
    naics_code = row['Most_Common_NAICS']
    avg_appv = row['Avg_Approval']
    avg_disb = row['Avg_Disbursement']
    avg_chgoff = row['Avg_ChargeOff']
    color = industry_colors.get(industry, '#000000')  # Set color, default to black if missing

    # check if the FG for this industry already exists; if not, create it
    if industry not in industry_groups:
        industry_groups[industry] = folium.FeatureGroup(name=industry).add_to(m)

    # create a GeoJson for the state and add it to the appropriate FG
    folium.GeoJson(
        state_geo,
        style_function=get_style_function(color),
        tooltip=f"State: {state_name} <br> Industry: {industry}",
        popup=folium.Popup(f"State: {state_name}<br>NAICS: {naics_code}<br>Industry: {industry}<br>Avg Approval: {avg_appv}<br>Avg Disbursement: {avg_disb}", max_width=300)
    ).add_to(industry_groups[industry])

    displayed_industries.add(industry)

# Create a custom legend control
legend_html = '''
<div style="position: fixed; 
            top: 50px; left: 50px; width: 195px; height: auto; 
            border:2px solid grey; z-index:9999; font-size:14px;
            background-color:white; opacity:0.8; padding: 10px;">
<h4 style="margin:0; padding-bottom:5px;">Industries</h4>
'''

for industry in displayed_industries:
    color = industry_colors.get(industry, '#000000')
    legend_html += f'<div><span style="display: inline-block; width: 10px; height: 10px; \
                         background-color: {color};"></span> {industry}</div>'
legend_html += '</div>'

# Add the custom legend to the map
legend = folium.Element(legend_html)
m.get_root().html.add_child(legend)

# Add LayerControl to the map
m.add_child(folium.LayerControl())

m
```

```python
import altair as alt

sampled_df = merged_industry.sample(n=1000, random_state=1)  # random_state ensures reproducibility

# Create a sample DataFrame without the complex data types
sampled_df_cleaned = sampled_df.drop(columns=['geometry'])  # Assuming 'geometry' is the column with Polygon objects

# Now, use this cleaned DataFrame for your Altair chart
# Ensure the DataFrame contains only the columns you'll use for the visualization
input_dropdown = alt.binding_select(options=['Industry', 'Region'])
selection = alt.selection_single(fields=['category'], bind=input_dropdown, name='View By')

# Base chart for charge-offs using the cleaned sampled DataFrame
base_chart = alt.Chart(sampled_df_cleaned).mark_bar().encode(
    x=alt.X('sum(ChgOffPrinGr):Q', title='Total Charge Off'),
    y=alt.Y('category:N', sort='-x'),
    color=alt.condition(selection, alt.Color('category:N', legend=None), alt.value('lightgray')),
    tooltip=[alt.Tooltip('category:N'), alt.Tooltip('sum(ChgOffPrinGr):Q', title='Total Charge Off')]
).add_params(
    selection
).transform_fold(
    fold=['Industry', 'Region'], as_=['category', 'value']
).transform_filter(
    selection
)

# Display the chart
base_chart.properties(width=600, height=400)
```
